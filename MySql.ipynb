# **SQL (Structured Query Language)**

SQL (Structured Query Language) is a standard programming language designed for managing and manipulating relational databases. It allows users to interact with databases to perform various operations like retrieving, inserting, updating, and deleting data.

**SQL is broadly divided into five categories based on the types of operations it performs:**

**DDL (Data Definition Language)**

DDL is used to define and manage database structures such as tables, schemas, and indexes.
Common Commands and Syntax

CREATE: Creates a new database object (table, view, etc.).



```
```CREATE TABLE table_name (
    column1 datatype,
    column2 datatype,
    ...
);




**ALTER: Modifies an existing database object**.



```
ALTER TABLE table_name
ADD column_name datatype;

```



**DROP: Deletes an existing database object.**

```
DROP TABLE table_name;

```

**TRUNCATE: Removes all records from a table without logging individual row deletions.**



```
TRUNCATE TABLE table_name;

```



# **DML (Data Manipulation Language)**

**DML is used to manipulate data within database tables.**

**INSERT: Adds new data to a table.**


```
INSERT INTO table_name (column1, column2, ...)
VALUES (value1, value2, ...);
```

**UPDATE: Modifies existing data in a table.**


```
UPDATE table_name
SET column1 = value1, column2 = value2
WHERE condition;

```


**DELETE: Removes data from a table.**


```
 DELETE FROM table_name
    WHERE condition;

```




# **DQL (Data Query Language)**

**DQL is used to query and retrieve data from the database.**

**SELECT: Retrieves data from one or more tables.**



```
SELECT column1, column2, ...
FROM table_name
WHERE condition;
```



# **DCL (Data Control Language)**

**DCL is used to control access to data in a database.**

**GRANT: Gives a user specific privileges.**

```
GRANT privilege_name
ON object_name
TO user_name;
```

**REVOKE: Removes specific privileges from a user.**



```
REVOKE privilege_name
ON object_name
FROM user_name;
```



# **TCL (Transaction Control Language)**

**TCL is used to manage transactions in a database, ensuring consistency and integrity.**

**COMMIT: Saves all changes made during the current transaction.**


```
COMMIT;

```


**ROLLBACK: Reverts changes made during the current transaction.**


```
ROLLBACK;
```


**SAVEPOINT: Creates a point within a transaction to which you can later rollback.**


```
SAVEPOINT savepoint_name;
```


**SET TRANSACTION: Sets the properties of the current transaction.**



```
SET TRANSACTION isolation_level;

```




**what is mean by ER Diagram?**

An ER (Entity-Relationship) Diagram is a visual tool used in database design to show how data is organized. It represents entities (like objects or people), their attributes (details about them), and the relationships between entities.

**Key Elements of an ER Diagram:**

**Entities:**
    
**These are objects or concepts in the real world that have data stored about them.**
        
Represented as rectangles.
        
Example: Student, Course, Library.

**Attributes:**
    
**These are the properties or details of an entity.**
        
Represented as ellipses (ovals).
        
Example: A Student entity might have attributes like Name, ID, and Age.

**Relationships:**
    
**These define how entities are related to each other.**
        
Represented as diamonds (rhombuses).
        
Example: A Student enrolls in a Course.

er2.jpg

# **What is mean by Triggers**

**a trigger is a database feature that automatically performs a specified action in response to certain events (like INSERT, UPDATE, or DELETE) on a table.**


```
CREATE TRIGGER trigger_name
{BEFORE | AFTER} {INSERT | UPDATE | DELETE}
ON table_name
FOR EACH ROW
BEGIN
    -- SQL statements to execute
END;

```


**Explanation:**

**CREATE TRIGGER:** Used to define a new trigger.
    
**trigger_name:** The name of the trigger.
    
**BEFORE or AFTER:** Specifies when the trigger is executed:
        
        BEFORE: The trigger executes before the event.
        
        AFTER: The trigger executes after the event.

**INSERT | UPDATE | DELETE:** Specifies the event that activates the trigger.

**table_name:** The table the trigger is attached to.
    
**FOR EACH ROW:** Ensures the trigger executes for each affected row.
    
**BEGIN ... END:** The block containing the SQL statements to execute.

# <b>Joins</b>
* used to combine rows from two or more tables, based on a related column between them(id,name)


# Types of Joins
* <u>INNER JOIN</u>: Returns records that have matching values in both tables
* <u>LEFT JOIN</u>: Returns all records from the left table, and the matched records from the right table
* <u>RIGHT JOIN</u>: Returns all records from the right table, and the matched records from the left table
* <u>CROSS JOIN</u>: Returns all records from both tables

j1.png

# Inner Join: "Selects records that have matching values in both tables"
        SELECT column_name(s)
        FROM table1
        INNER JOIN table2
        ON table1.column_name = table2.column_name;

in.jpeg








# LEFT JOIN: "Returns all records from the left table1, and the matching records (if any) from the right table2"
        SELECT column_name(s)
        FROM table1
        LEFT JOIN table2
        ON table1.column_name = table2.column_name;


lf.jpeg






# RIGHT JOIN: "Returns all records from the right table2, and the matching records (if any) from the left table1"
        SELECT column_name(s)
        FROM table1
        RIGHT JOIN table2
        ON table1.column_name = table2.column_name;


rt.jpeg

# Full Join:Returns all records when there is a match in either Left or Right table.


NOTE:IN mysql we can't use FULL JOIN keyword if you want to perform full join we can uses UNION keyword(return unique values).
```

```

ful.jpeg



# CROSS JOIN: "Returns all records from both tables"
        SELECT column_name(s)
        FROM table1
        CROSS JOIN table2;
# Self Join: "Table is joined with itself"
        SELECT column_name(s)
        FROM table11
        JOIN table12
        ON table1.col_name=table2.col_name;
self.jpeg





# <b>UNION</b>
* used to combine the result-set of two or more SELECT statements
* Every SELECT statement within UNION must have the same number of columns
* The columns must also have similar data types
* The columns in every SELECT statement must also be in the same order

# Note:**Union** is used to remove the duplications where **union ALL** is allow the duplications.



# Union:
        SELECT column_name(s) FROM table1
        UNION
        SELECT column_name(s) FROM table2;
        ORDER BY columan_name; //Optional
# Union All:
        SELECT column_name(s) FROM table1
        UNION ALL
        SELECT column_name(s) FROM table2
        ORDER BY columan_name; //Optional
# Union With Where:
        SELECT column_name(s) FROM table1
        WHERE column_name = 'value1'
        UNION
        SELECT column_name(s) FROM table2
        WHERE column_name = 'value1'
        ORDER BY columan_name; //Optional
# Union All With Where:
        SELECT column_name(s) FROM table1
        WHERE column_name = 'value1'
        UNION All
        SELECT column_name(s) FROM table2
        WHERE column_name = 'value1'
        ORDER BY columan_name; //Optional

# **what is mean by sub query**

**A subquery in SQL is a query nested inside another query. It is used to perform an operation that provides data for the main query.**

Sc.png

sub.jpeg

# <b>GROUP BY</b>
* The GROUP BY statement is often used with aggregate functions (COUNT(), MAX(), MIN(), SUM(), AVG()) to group the result-set by one or more columns.

        SELECT column_name(s)
        FROM table_name
        WHERE condition
        GROUP BY column_name(s)
        ORDER BY column_name(s);

# <b>HAVING</b>
* The HAVING clause was added to SQL because the WHERE keyword cannot be used with aggregate functions

        SELECT column_name(s)
        FROM table_name
        WHERE condition
        GROUP BY column_name(s)
        HAVING condition
        ORDER BY column_name(s);

# <b>EXISTS</b>
* Used to test for the existence of any record in a subquery
* Returns TRUE if the subquery returns one or more records

        SELECT column_name(s)
        FROM table_name
        WHERE EXISTS (SELECT column_name FROM table_name WHERE condition);

# <b>ANY and ALL</b>
* The ANY and ALL operators allow you to perform a comparison between a single column value and a range of other values

# Any:
* Returns a boolean value as a result
* ANY means that the condition will be true if the operation is true for any of the values in the range.

        SELECT column_name(s)
        FROM table_name
        WHERE column_name operator ANY (SELECT column_name FROM table_name WHERE condition);


# All:
* Returns a boolean value as a result
* Returns TRUE if ALL of the subquery values meet the condition

# ALL Syntax With SELECT:
        SELECT ALL column_name(s)
        FROM table_name
        WHERE condition;
# ALL Syntax With WHERE or HAVING:
        SELECT column_name(s)
        FROM table_name
        WHERE column_name operator ALL (SELECT column_name FROM table_name WHERE condition);

# <b>CASE</b>
* The CASE statement goes through conditions and returns a value when the first condition is met (like an if-then-else statement). So, once a condition is true, it will stop reading and return the result. If no conditions are true, it returns the value in the ELSE clause.
* If there is no ELSE part and no conditions are true, it returns NULL.

        SELECT column1, colummn2,
        CASE
            WHEN condition1 THEN result1
            WHEN condition2 THEN result2
            WHEN conditionN THEN resultN
            ELSE result
        END;
        FROM Database_name;

# <center><b><u>MySQL Operators</u></b></center>

# Arithmetic Operators
* '+' ->	Add
* '-' ->	Subtract
* '*' ->	Multiply
* '/' ->	Divide
* '%' ->	Modulo

# Bitwise Operators
* '&'	Bitwise AND
* '|'	Bitwise OR
* '^'	Bitwise exclusive OR

# Comparison Operators
* '=' ->	Equal to
* '>' ->	Greater than
* '<' ->	Less than
* '>=' ->	Greater than or equal to
* '<=' ->	Less than or equal to
* '<>' ->	Not equal to

# Compound Operators
* '+=' ->	Add equals
* '-=' ->	Subtract equals
* '*=' ->	Multiply equals
* '/=' ->	Divide equals
* '%=' ->	Modulo equals
* '&=' ->	Bitwise AND equals
* '^-=' ->	Bitwise exclusive equals
* '|*=' ->	Bitwise OR equals

# Logical Operators
* 'ALL' ->	TRUE if all of the subquery values meet the condition
* 'AND' ->	TRUE if all the conditions separated by AND is TRUE
* 'ANY' ->	TRUE if any of the subquery values meet the condition
* 'BETWEEN' ->	TRUE if the operand is within the range of comparisons
* 'EXISTS' ->	TRUE if the subquery returns one or more records
* 'IN' ->	TRUE if the operand is equal to one of a list of expressions
* 'LIKE' ->	TRUE if the operand matches a pattern
* 'NOT' ->	Displays a record if the condition(s) is NOT TRUE
* 'OR' ->	TRUE if any of the conditions separated by OR is TRUE
* 'SOME' ->	TRUE if any of the subquery values meet the condition

# <b>ALTER TABLE</b>
* used to add, delete, or modify columns in an existing table
* And used to add and drop various constraints on an existing table

# 1. ADD Column: "To add a column in a table"
        ALTER TABLE table_name
        ADD column_name datatype;
# 2. DROP COLUMN: "To delete a column in a table"
        ALTER TABLE table_name
        DROP COLUMN column_name;
# 3. MODIFY COLUMN: "To change the data type of a column in a table"
        ALTER TABLE table_name
        MODIFY COLUMN column_name datatype;

# <b>Constraints</b>
* used to specify rules for data in a table
* used to limit the type of data that can go into a table

The following constraints are commonly used in SQL:
* <u>NOT NULL</u> - Ensures that a column cannot have a NULL value
* <u>UNIQUE</u> - Ensures that all values in a column are different
* <u>PRIMARY KEY</u> - A combination of a <mark style = 'color:#fff; background-color:#DC143C'>NOT NULL</mark> and <mark style = 'color:#fff; background-color:#DC143C'>UNIQUE</mark>. Uniquely identifies each row in a table
* <u>FOREIGN KEY</u> - Prevents actions that would destroy links between tables
* <u>CHECK</u> - Ensures that the values in a column satisfies a specific condition
* <u>DEFAULT</u> - Sets a default value for a column if no value is specified
* <u>CREATE INDEX</u> - Used to create and retrieve data from the database very quickly

<u>Syntax</u>:

        CREATE TABLE table_name (
            column1 datatype constraint,
            column2 datatype constraint,
            column3 datatype constraint,
            ....
        );

# <b>NOT NULL</b>
# NOT NULL on CREATE TABLE
        CREATE TABLE Persons (
            ID int NOT NULL,
            LastName varchar(255) NOT NULL,
            FirstName varchar(255) NOT NULL,
            Age int
        );
# NOT NULL on ALTER TABLE
        ALTER TABLE Persons
        MODIFY Age int NOT NULL;

# <b>UNIQUE</b>
# UNIQUE Constraint on CREATE TABLE
        CREATE TABLE Persons (
        ID int NOT NULL,
        LastName varchar(255) NOT NULL,
        FirstName varchar(255),
        Age int,
        UNIQUE (ID)
        );
# UNIQUE Constraint on ALTER TABLE
        ALTER TABLE Persons
        ADD UNIQUE (ID);
# DROP a UNIQUE Constraint
        ALTER TABLE Persons
        DROP INDEX UC_Person;

# <b>PRIMARY KEY</b>
# PRIMARY KEY on CREATE TABLE
        CREATE TABLE Persons (
            ID int NOT NULL,
            LastName varchar(255) NOT NULL,
            FirstName varchar(255),
            Age int,
            PRIMARY KEY (ID)
        );
# PRIMARY KEY on ALTER TABLE
        ALTER TABLE Persons
        ADD PRIMARY KEY (ID);
# DROP a PRIMARY KEY Constraint
        ALTER TABLE Persons
        DROP PRIMARY KEY;

# <b>FOREIGN KEY</b>
# FOREIGN KEY on CREATE TABLE
        CREATE TABLE Orders (
            OrderID int NOT NULL,
            OrderNumber int NOT NULL,
            PersonID int,
            PRIMARY KEY (OrderID),
            FOREIGN KEY (PersonID) REFERENCES Persons(PersonID)
        );
# FOREIGN KEY on ALTER TABLE
        ALTER TABLE Orders
        ADD FOREIGN KEY (PersonID) REFERENCES Persons(PersonID);
# DROP a FOREIGN KEY Constraint
        ALTER TABLE Orders
        DROP FOREIGN KEY FK_PersonOrder;

# <b>CHECK</b>
# CHECK on CREATE TABLE
        CREATE TABLE Persons (
            ID int NOT NULL,
            LastName varchar(255) NOT NULL,
            FirstName varchar(255),
            Age int,
            CHECK (Age>=18)
        );
# CHECK on ALTER TABLE
        ALTER TABLE Persons
        ADD CHECK (Age>=18);
# DROP a CHECK Constraint
        ALTER TABLE Persons
        DROP CHECK CHK_PersonAge;

# <b>DEFAULT</b>
# DEFAULT on CREATE TABLE
        CREATE TABLE Persons (
            ID int NOT NULL,
            LastName varchar(255) NOT NULL,
            FirstName varchar(255),
            Age int,
            City varchar(255) DEFAULT 'Sandnes'
        );
# DEFAULT on ALTER TABLE
        ALTER TABLE Persons
        ALTER City SET DEFAULT 'Sandnes';
# DROP a DEFAULT Constraint
        ALTER TABLE Persons
        ALTER City DROP DEFAULT;

# <b>CREATE INDEX</b>
# CREATE INDEX
        CREATE INDEX index_name
        ON table_name (column1, column2, ...);
# CREATE UNIQUE INDEX
        CREATE UNIQUE INDEX index_name
        ON table_name (column1, column2, ...);
# DROP INDEX
        ALTER TABLE table_name
        DROP INDEX index_name;

# <b>AUTO INCREMENT</b>
* Auto-increment allows a unique number to be generated automatically when a new record is inserted into a table.

* Often this is the primary key field that we would like to be created automatically every time a new record is inserted.

        CREATE TABLE Persons (
            Personid int NOT NULL AUTO_INCREMENT,
            LastName varchar(255) NOT NULL,
            FirstName varchar(255),
            Age int,
            PRIMARY KEY (Personid)
        );

# <b>Working With Dates</b>
# MySQL Date Data Types
MySQL comes with the following data types for storing a date or a date/time value in the database:

* <u>DATE</u> -> format YYYY-MM-DD
* <u>DATETIME</u> -> format: YYYY-MM-DD HH:MI:SS
* <u>TIMESTAMP</u> -> format: YYYY-MM-DD HH:MI:SS
* <u>YEAR</u> -> format YYYY or YY

        SELECT * FROM Orders WHERE OrderDate='2008-11-11'

# <b>CREATE VIEW</b>
* A view contains rows and columns, just like a real table. The fields in a view are fields from one or more real tables in the database.

 views in SQL are not temporary data. A view is a stored query that acts like a virtual table. It is based on the result set of an SQL query and is permanently saved in the database schema.

# Creating a View
        CREATE VIEW view_name AS
        SELECT column1, column2, ...
        FROM table_name
        WHERE condition;
# Updating a View
        CREATE OR REPLACE VIEW view_name AS
        SELECT column1, column2, ...
        FROM table_name
        WHERE condition;
# Dropping a View
        DROP VIEW view_name;

# **Normalization**

**Normalization is the process of organizing data in a database to avoid duplication and make it efficient.**

**Types of Normalization**

**First Normal Form (1NF):**

Ensures that all columns contain atomic (indivisible) values.
    
Eliminates duplicate columns from the same table.
    
Example: Splitting a list of multiple phone numbers in a single cell into separate rows or columns.

1nf.png

# **Second Normal Form (2NF)**

Second Normal Form (2NF) is a level of database normalization used to organize data in a relational database to reduce redundancy and improve data integrity. A table is in 2NF if it satisfies the following conditions:

**It is in First Normal Form (1NF):**
        
The table must have atomic values (indivisible values in each cell).
        
No repeating groups or multi-valued attributes are allowed.
        
Each row is unique and identifiable by a primary key.

**It has no Partial Dependencies:**
        
All non-key attributes must depend on the entire primary key, not just part of it.
        
If the primary key is a composite key (made up of two or more columns), no non-key attribute should depend on only part of this key.
        
If the primary key is a single column, partial dependency is not possible, and the table already satisfies this condition.

**Step 3: Checking for 2NF (Identifying Partial Dependencies)**

Primary Key  

If we assume the composite primary key is (CustomerID, PhoneNumber):

CustomerName and City depend only on CustomerID (not on the full composite key).
    
This is a partial dependency, violating 2NF.

2NF.png

# **3NF Definition and Key Points:**

**3NF Builds on 2NF:**
        
A table must first satisfy Second Normal Form (2NF) (i.e., no partial dependencies).

**No Transitive Dependencies:**
        
There should be no transitive dependency, where a non-key attribute depends on another non-key attribute.
        
All non-key attributes should depend only on the primary key.

**Key Rule**:
        
A table is in 3NF if:
        
Every non-key attribute is functionally dependent on the whole primary key and nothing else.

**Step 2: Introducing Transitive Dependencies**

Suppose the City in the Customers Table has a related attribute, Region (e.g., "New York" belongs to the "East" region, "Chicago" to the "Midwest," etc.). The new Customers Table might look like this:

333.png

3333.png

# **Boyce-Codd Normal Form (BCNF)**

**Boyce-Codd Normal Form (BCNF) is a stricter version of Third Normal Form (3NF). A table is in BCNF if:**

It is in 3NF.
    
For every functional dependency, the left-hand side (determinant) must be a superkey.(X->Y WHERE X is a super key)

In simpler terms, BCNF removes any remaining anomalies by ensuring that every non-trivial functional dependency has a determinant that is a superkey

bcnf.png

#**Fourth Normal Form (4NF)**

Fourth Normal Form (4NF) addresses a specific type of redundancy called multivalued dependencies. A table is in 4NF if:

It is in Boyce-Codd Normal Form (BCNF).
    
It does not have any multivalued dependencies.

**What is a Multivalued Dependency?**

A multivalued dependency occurs when one attribute determines a set of values for another attribute (or set of attributes) that are independent of each other. This situation typically arises when one attribute is related to more than one value, and those values don't depend on each other.

For example, if a table has an attribute that can have multiple values, and those values are independent of one another, then we have a multivalued dependency.


4nf.png

4444.png

# **Fifth Normal Form (5NF)**

**Relation in 4NF:**
        
The table should already be in Fourth Normal Form (4NF), meaning there are no multivalued dependencies (e.g., independent lists in a single table).

**Avoid Join Dependency Anomalies:**
        
When a table is split into smaller ones, it shouldn't create anomalies where:
  
Data becomes inconsistent.
          
Joining tables leads to duplicate or incorrect rows.

**Lossless Decomposition:**
        
After splitting, joining the smaller tables should recreate the original table perfectly, without losing any data.



5nf.png

# **Step 2: Decompose the Table**

**Decomposition into 5NF:**


5nf2.png

# **Step 3: Verify Lossless Decomposition**

Join Customer-PhoneNumber with PhoneNumber-Service on PhoneNumber. This gives a table mapping each CustomerID to PhoneNumber and Service.
    
Join the resulting table with Customer-Service on CustomerID. This reconstructs the original table without any loss of data.

5nf3.png

# **Transaction in MySQL**

**A transaction in MySQL is a sequence of one or more SQL operations (such as INSERT, UPDATE, or DELETE) that are treated as a single unit of work. Transactions ensure that these operations are executed in a reliable, consistent, and atomic way.**

**Key Properties of Transactions (ACID)**

**Atomicity:**
        
Ensures that all operations in a transaction are completed. If one operation fails, the entire transaction is rolled back, leaving the database unchanged.

**Consistency:**
        
Ensures that the database remains in a valid state before and after the transaction.

**Isolation:**
        
Ensures that transactions are executed independently of one another, preventing conflicts caused by simultaneous operations.

**Durability:**
        
Ensures that once a transaction is committed, the changes are permanent, even in the event of a system failure.

**if you want to transfer the money from one bank to other bank.**

a = read(AccountA_balance);

a = a - 50;

write(AccountA_balance, a);

b = read(AccountB_balance);

b = b + 50;

write(AccountB_balance, b);


# **Data abstraction**

**Data abstraction in MySQL means hiding the complex details of how data is stored and managed and showing users only what they need to know to work with the data.**

**It has three levels:**

Physical Level: How data is stored on the server (e.g., files on disk).
    
Logical Level: How the data is organized into tables, columns, and relationships (e.g., database schema).
    
View Level: What the user sees, such as specific data through queries or views.

**Example:**

Physical Level: Data is stored in files like users.ibd.
    
Logical Level: You define a table like:

**CREATE TABLE users (id INT, name VARCHAR(50));**

View Level: You can create a view to show only specific data:

**CREATE VIEW user_names AS SELECT name FROM users;**
